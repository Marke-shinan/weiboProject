## 评论数据处理

```python
import re
col_name='评论内容'
df_data[col_name]=df_data[col_name].astype('str')
def dealContent(x):
    """正则表达式删除固定句式"""
    x=re.sub('（.*?）|\(.*?\)|【.*?】|@回复 |转发.*','',x)
    x=re.sub('#.*?#|http.*://@.*?:|「.*?」|@.*? |『.*?』|“.*?”|《.*?》|offer.*|原创来自：.*','',x)
    x=x.replace('nan','')
    if x=='' or x==' ':
        x=None
    return x
df_data[col_name]=df_data[col_name].apply(lambda x:dealContent(x))
df_data.dropna(subset=[col_name],inplace=True)
df_data.reset_index(drop=True,inplace=True)

def dealCommentLen(x):
    """正则表达式提取中文信息"""
    x = re.sub("[^\u4e00-\u9fa5]+", " ", x) # 只保留中文 
    x_=x.replace(' ','')
    if x_=='' or x_==' '  or x_==None:
        return None
    if len(x_)==0:
        return None
    return x
df_data[col_name+"clear"]=df_data[col_name].apply(lambda x:dealCommentLen(x))
df_data.dropna(subset=[col_name+"clear"],inplace=True)
df_data.reset_index(drop=True,inplace=True)
df_data
```

# 主题分析

## 主题分析代码

```python
def ldaAnalysis(data_set,num_topics=3, topn=10,saveName=''):
    """
        :param comment_data: dataframe数据
        :param num_topics: 主题数量
        :param topn: topn设置词语数量
        :return:
    """
    data_dict = corpora.Dictionary(data_set)  # 建立词典 # 构建 document-term matrix
    # [[(0, 1), (1, 1), (2, 1), (3, 1), (4, 1), (5, 1)]]
    corpus = [data_dict.doc2bow(i) for i in data_set]  # 建立语料库
    pos_corpus_idf = models.TfidfModel(corpus)[corpus]
    lda_model = models.LdaModel(pos_corpus_idf,iterations=200, num_topics=num_topics, id2word=data_dict)  # LDA模型训练
    with open(f"data/lda主题{saveName}.txt", mode='w', encoding='utf-8-sig') as f:
        for i in range(num_topics):
            print(f'\033[5;33;5m主题:--->{str(i + 1)}\033[0m')
            print("\t", lda_model.print_topic(i, topn=topn))  # 输出每个主题 topn设置词语数量
            f.write(lda_model.print_topic(i, topn=topn))
            f.write('\n')
    # 打印所有主题，每个主题显示10个词
    # 用pyLDAvis将LDA模式可视化
    plot = pyLDAvis.gensim_models.prepare(lda_model, corpus, data_dict)
    # 保存到本地html
    pyLDAvis.save_html(plot, f'data/img_LDA_pyLDAvis{saveName}.html')
#     return data_dict, corpus, pos[2]
```

```python
from scipy.linalg.special_matrices import triu
import codecs
import re
import pandas as pd
import jieba
import matplotlib
import pyLDAvis.gensim_models
import matplotlib.pyplot as plt
from gensim import corpora, models
import jieba#用来分词，也可以进行词性标注
import re#正则表达式清洗数据
from collections import Counter#统计词频
from gensim.models.ldamodel import LdaModel
#此处要去除停用词两边的空格和换行符
import codecs#用户加载停用词，直接open是乱码，这个是转化为unicode编码
stopword = codecs.open('data/data_stopwords.txt', 'r',encoding='utf-8')  # 加载停用词
stopword=[word.strip() for word in stopword]
from gensim.models.coherencemodel import CoherenceModel


#计算coherence
def coherence(num_topics,corpus,dictionary,data_set):
    print('================= coherence =================')
    print(f'主题数量: {num_topics}')
    ldamodel = LdaModel(corpus, num_topics=num_topics, id2word = dictionary, passes=30,random_state = 1)
    print("主题词: ",ldamodel.print_topics(num_topics=num_topics, num_words=10)[num_topics-1])
    ldacm = CoherenceModel(model=ldamodel, texts=data_set, dictionary=dictionary, coherence='c_v')
    print("coherence= ",ldacm.get_coherence())
    return ldacm.get_coherence()
    

def calculateCoherence(data_set):# LDA 一致性检验
    """
        :param comment_data: dataframe数据
        :param topn: topn设置词语数量
        :return:
    """
    dictionary = corpora.Dictionary(data_set)  # 建立词典
    # [[(0, 1), (1, 1), (2, 1), (3, 1), (4, 1), (5, 1)]]
    corpus = [dictionary.doc2bow(i) for i in data_set]  # 建立语料库
    # 绘制主题-coherence曲线，选择最佳主题数
    plt.figure(figsize=(8,8),dpi=100)
    x = range(1,6)
    # z = [perplexity(i) for i in x]  #如果想用困惑度就选这个
    y=[]
    for i in x:
        score = coherence(i,corpus,dictionary,data_set) 
        y.append(score)
        plt.text(i, score, f'({i},{round(float(score),3)})', fontsize=10)
    plt.plot(x, y)
    plt.xlabel('主题数目')
    plt.ylabel('coherence大小')
    plt.title('主题-coherence变化情况')
    plt.savefig('data/img_LDA_coherence.png')
    plt.show()
#     return dictionary, corpus,data_set
    
```

## 主题分析-普通用户

```python
col_info='普通用户'
all_bowen=df_data0[df_data0['微博认证']==col_info]['博文内容clear'].to_list() + df_data[df_data['微博认证']==col_info]['评论内容clear'].to_list()
```

```python
df=pd.DataFrame()
df['数据']=all_bowen
```

```python
def jieba_cuts(commentss,stopword):
    """
    commentss:待分词的字符串
    stopword:停用词
    """
    line = re.sub("[^\u4e00-\u9fa5]+", " ", commentss)#正则表达式只保留中文
    all_words=jieba.cut(line,cut_all=False)#结巴分词，三种模式，此处是精确模式，另外两种分别是：全模式和搜索模式
    c=[]
    for x in all_words:#遍历词汇进行统计，只保留分词后的词大于单个字符，去除停用词
        x=x.replace(' ','')
        if len(x)>0 and x != '\r\n'  and x not in stopword: #len(x)>1 and 
            c.append(x)
    return c


colName='数据'
# 将评论数据变成字符串，防止空值的影响
df[colName] = df[colName].astype('str')
mycut = lambda x: jieba_cuts(x,stopword)  # 自定义简单分词函数
df['corp'] = df[colName].apply(mycut) # 将每个文档分词成单词列表
# df.to_excel(f'data/data_{col_info}.xlsx')

col_info='普通用户'
aaa=df_data0[df_data0['微博认证']==col_info][['博文发布者名称','微博认证','最后评论时间','博文内容']]
bbb=df_data[df_data['微博认证']==col_info][['评论用户名','微博认证','评论时间','评论内容']]
aaa.columns=['发布者名称','微博认证','评论时间','内容']
aaa['flag']='博文'
bbb.columns=['发布者名称','微博认证','评论时间','内容']
bbb['flag']='评论'
df_save=pd.concat([aaa,bbb])
df_save.to_excel(f'data/data_{col_info}.xlsx',index=False)
```

```python
# df_pos=df_pos['corp'][:100]#建立存储分词的列表
calculateCoherence(data_set=df['corp']) # 调用calculateCoherence函数，用于计算一致性
```
